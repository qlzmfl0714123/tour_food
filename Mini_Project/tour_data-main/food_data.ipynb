```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv("data/맛집.csv", skiprows=3, encoding='utf-8')  # 또는 encoding='cp949' / 'euc-kr' 시도

# 데이터 확인
print(df.head())
```

```python
df.columns = ['지역', '도시명', '음식종류', '대표메뉴', '식당상호', '포털 검색', '추천사유']
df
```

```python
df['지역'].value_counts()
```

```python
gang_df = df.loc[df['지역'] == '강원도']

gang_df['도시명'].value_counts()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import seaborn as sns
```

```python
font_path = 'C:\\Windows\\Fonts\\malgun.ttf'

font_prop = fm.FontProperties(fname=font_path).get_name()
print(font_prop)

matplotlib.rc('font', family=font_prop)
```

```python
figure, axes1 = plt.subplots(nrows=1, ncols=1)
figure.set_size_inches(18,12)

sns.countplot(data=gang_df, x='도시명', ax=axes1, order=gang_df['도시명'].value_counts().index, hue='도시명')

plt.title("강원도지역의 도시별 맛집 수")
plt.show()
```

```python
food_kind = gang_df['음식종류'].value_counts(normalize=True)
food_kind
```

```python
figure = plt.figure(figsize=(20,12))

plt.pie(food_kind, labels=food_kind.index, autopct='%1.1f%%', startangle=95,shadow=True)

plt.axis('equal')
plt.title("강원도지역의 음식종류 맛집 수")
plt.show()
```
